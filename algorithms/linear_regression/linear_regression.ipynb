```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
from linear_regression import LinearRegression
```

```python
boston_dataset = pd.read_csv(
    "https://raw.githubusercontent.com/AbrLabs/datasets/master/BostonHousing.csv"
)
x_boston = boston_dataset.iloc[:, :-1].values
y_boston = boston_dataset.iloc[:, -1].values
```

```python
pca_boston = PCA(n_components=1)
x_boston = pca_boston.fit_transform(x_boston)
x_boston = StandardScaler().fit_transform(x_boston)
x_boston_train, x_boston_test, y_boston_train, y_boston_test = train_test_split(
    x_boston, y_boston, test_size=0.2
)
```

```python
model_LR_boston = LinearRegression()
model_LR_boston.fit(x_boston_train, y_boston_train)
```

```python
plt.scatter(x_boston_train, y_boston_train, color="red")
plt.plot(x_boston_train, model_LR_boston.predict(x_boston_train), color="blue")
plt.title("Boston Housing Prices")
plt.xlabel("PCA")
plt.ylabel("Price")
plt.show()
```
